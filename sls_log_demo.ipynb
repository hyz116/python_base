```python
import json
# 云日志解析处理
# 解析从阿里云下载下来的sls 日志，取得key ['content']中的内容，读取hid 和 rp id

## 读取文件，一行一行的读取
with open('/Users/chuye/Documents/Alibaba/aliyun/slsLog/downloaded_data.txt', 'r') as fin:
    lines = fin.readlines()
for line in lines:
    # to json 
    line_josn = json.loads(line)
    # 以逗号分隔
    content_str = line_josn['content']
    sub_context_list = content_str.split(',')
    print('hid:{},rpId:{}\n'.format(sub_context_list[1].replace('hid:',''), sub_context_list[2].replace('RatePlanId','')))
```
